```python
NAMES=[]
name = driver.find_elements_by_xpath("//h3[@class='product-brand']")
time.sleep(5)
if name is not None:
    for i in name:
        name1=i.text
        NAMES.append(name1)
    else:
        NAMES.append('-')
```

```python
NAMES
```

```python
PRICE=[]
price = driver.find_elements_by_xpath("//span[@class='product-discountedPrice']")
time.sleep(5)
if price is not None:
    for i in price:
        name1=i.text
        PRICE.append(name1)
    else:
        PRICE.append('-')
```

```python
PRICE
```

```python
DISCRIPTION=[]
discription = driver.find_elements_by_xpath("//h4[@class='product-product']")
time.sleep(5)
if discription is not None:
    for i in discription:
        name1=i.text
        DISCRIPTION.append(name1)
    else:
        DISCRIPTION.append('-')
```

```python
DISCRIPTION
```

```python
MYNTRASHOES=pd.DataFrame({'NAMES': pd.Series(NAMES), 'PRICE': pd.Series(PRICE), 'DISCRIPTION': pd.Series(DISCRIPTION)})
```
